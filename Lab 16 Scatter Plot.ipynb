```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
conn = sqlite3.connect('survey-data.sqlite')
```


### Task 1: Exploring Relationships with Scatter Plots



#### 1. Scatter Plot for Age vs. Job Satisfaction



Visualize the relationship between respondents' age (`Age`) and job satisfaction (`JobSatPoints_6`). Use this plot to identify any patterns or trends.




```python
## Write your code here

# Query the data
query = """
SELECT Age, JobSatPoints_6
FROM main
WHERE Age IS NOT NULL AND JobSatPoints_6 IS NOT NULL
"""
df = pd.read_sql_query(query, conn)

# Ensure 'Age' is a valid categorical column and 'JobSatPoints_6' is numeric
df['JobSatPoints_6'] = pd.to_numeric(df['JobSatPoints_6'], errors='coerce')

# Drop rows with missing or invalid values in Age or JobSatPoints_6
df = df.dropna(subset=['Age', 'JobSatPoints_6'])

# Create the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['JobSatPoints_6'], alpha=0.5, c='blue', edgecolors='w', s=70)

# Customize the plot
plt.title('Scatter Plot: Age vs. Job Satisfaction')
plt.xlabel('Age')
plt.xticks(rotation=90, ha='right')
plt.ylabel('Job Satisfaction (JobSatPoints_6)')
plt.grid(True)

# Show the plot
plt.show()
```

#### 2. Scatter Plot for Compensation vs. Job Satisfaction


Explore the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`) using a scatter plot.


```python
## Write your code here
# Query the data
query = """
SELECT ConvertedCompYearly, JobSatPoints_6
FROM main
WHERE ConvertedCompYearly IS NOT NULL AND JobSatPoints_6 IS NOT NULL
"""
df = pd.read_sql_query(query, conn)


# Drop rows with missing or invalid values
df = df.dropna(subset=['ConvertedCompYearly', 'JobSatPoints_6'])

# Create the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['ConvertedCompYearly'], df['JobSatPoints_6'], alpha=0.5, c='green', edgecolors='w', s=70)

# Customize the plot
plt.title('Scatter Plot: Compensation vs. Job Satisfaction')
plt.xlabel('Yearly Compensation (ConvertedCompYearly)')
plt.ylabel('Job Satisfaction (JobSatPoints_6)')
plt.grid(True)

# Show the plot
plt.show()
```

### Task 2: Enhancing Scatter Plots


#### 1. Scatter Plot with Trend Line for Age vs. Job Satisfaction



Add a regression line to the scatter plot of Age vs. JobSatPoints_6 to highlight trends in the data.


```python
cursor = conn.cursor()

# Step 1: Add a new REAL column
cursor.execute("ALTER TABLE main ADD COLUMN Age_Real REAL;")

# Step 2: Update the new column with numeric values
cursor.execute("""
UPDATE main
SET Age_Real = 
    CASE 
        WHEN Age = '18-24 years old' THEN 21.0
        WHEN Age = '25-34 years old' THEN 29.5
        WHEN Age = '35-44 years old' THEN 39.5
        WHEN Age = '45-54 years old' THEN 49.5
        WHEN Age = '55-64 years old' THEN 59.5
        WHEN Age = '65 years or older' THEN 70.0
        WHEN Age = 'Prefer not to say' THEN NULL
        WHEN Age = 'Under 18 years old' THEN 16.0
    END;
""")

# Commit changes and close the connection
conn.commit()
```

```python
## Write your code here

# Query the data
query = """
SELECT Age_Real, JobSatPoints_6
FROM main
WHERE Age_Real IS NOT NULL AND JobSatPoints_6 IS NOT NULL
"""
df = pd.read_sql_query(query, conn)

# Ensure 'JobSatPoints_6' is numeric and 'Age' is a valid categorical column
df['JobSatPoints_6'] = pd.to_numeric(df['JobSatPoints_6'], errors='coerce')

# Drop rows with missing or invalid values
df = df.dropna(subset=['Age_Real', 'JobSatPoints_6'])

# Create the scatter plot with regression line
plt.figure(figsize=(10, 6))
sns.regplot(x='Age_Real', y='JobSatPoints_6', data=df, scatter_kws={'alpha':0.5, 'color':'blue'}, line_kws={'color':'red', 'linewidth':2})

# Customize the plot
plt.title('Scatter Plot with Regression Line: Age vs. Job Satisfaction')
plt.xlabel('Age')
plt.ylabel('Job Satisfaction (JobSatPoints_6)')
plt.grid(True)

# Show the plot
plt.show()
```

### Task 3: Combining Scatter Plots with Additional Features


#### 1. Bubble Plot of Compensation vs. Job Satisfaction with Age as Bubble Size



Create a bubble plot to explore the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`), with bubble size representing age.


```python
## Write your code here

# Query the data
query = """
SELECT ConvertedCompYearly, JobSatPoints_6, Age_Real
FROM main
WHERE ConvertedCompYearly IS NOT NULL AND JobSatPoints_6 IS NOT NULL AND Age IS NOT NULL
"""
df = pd.read_sql_query(query, conn)

# Drop rows with missing or invalid values
df = df.dropna(subset=['ConvertedCompYearly', 'JobSatPoints_6', 'Age_Real'])

# Create the bubble plot
plt.figure(figsize=(12, 8))
plt.scatter(
    df['ConvertedCompYearly'], df['JobSatPoints_6'],
    s=df['Age_Real'] * 10,  # Bubble size is proportional to Age
    alpha=0.6,         # Transparency for bubbles
    c=df['JobSatPoints_6'],  # Color the bubbles based on JobSatPoints_6
    cmap='viridis',    # Color map
    edgecolors='w',    # White border around the bubbles
)

# Customize the plot
plt.title('Bubble Plot: Compensation vs. Job Satisfaction with Age as Bubble Size')
plt.xlabel('Yearly Compensation (ConvertedCompYearly)')
plt.ylabel('Job Satisfaction (JobSatPoints_6)')
plt.colorbar(label='Job Satisfaction')  # Show color bar
plt.grid(True)

# Show the plot
plt.show()
```

#### 2. Scatter Plot for Popular Programming Languages by Job Satisfaction


Visualize the popularity of programming languages (`LanguageHaveWorkedWith`) against job satisfaction using a scatter plot. Use points to represent satisfaction levels for each language.


```python
## Write your code here

# Query the data
query = """
SELECT LanguageHaveWorkedWith, JobSatPoints_6
FROM main
WHERE LanguageHaveWorkedWith IS NOT NULL AND JobSatPoints_6 IS NOT NULL
"""
df = pd.read_sql_query(query, conn)

# Split 'LanguageHaveWorkedWith' into individual languages
df['LanguageHaveWorkedWith'] = df['LanguageHaveWorkedWith'].str.split(',')

# Explode the list of languages so each language gets its own row
languages_expanded = df.explode('LanguageHaveWorkedWith')

# Remove leading/trailing spaces from language names
languages_expanded['LanguageHaveWorkedWith'] = languages_expanded['LanguageHaveWorkedWith'].str.strip()

# Convert JobSatPoints_6 to numeric
languages_expanded['JobSatPoints_6'] = pd.to_numeric(languages_expanded['JobSatPoints_6'], errors='coerce')

# Group by Language and calculate metrics
language_stats = languages_expanded.groupby('LanguageHaveWorkedWith').agg(
    popularity=('JobSatPoints_6', 'size'),  # Count the number of respondents
    avg_job_satisfaction=('JobSatPoints_6', 'mean')  # Calculate average job satisfaction
).reset_index()

# Drop rows with missing values
language_stats = language_stats.dropna()

# Create scatter plot
plt.figure(figsize=(20,20))
plt.scatter(
    language_stats['popularity'], language_stats['avg_job_satisfaction'],
    alpha=0.8, s=100, c=language_stats['avg_job_satisfaction'], cmap='viridis', edgecolors='w'
)

# Customize plot
plt.title('Scatter Plot: Popularity of Programming Languages vs. Job Satisfaction')
plt.xlabel('Popularity (Number of Respondents)')
plt.ylabel('Average Job Satisfaction')
plt.colorbar(label='Job Satisfaction')
plt.grid(True)

# Annotate top languages
top_languages = language_stats.sort_values('popularity', ascending=False).head(5)
for _, row in top_languages.iterrows():
    plt.text(
        row['popularity'], row['avg_job_satisfaction'], row['LanguageHaveWorkedWith'],
        fontsize=10, ha='right', va='bottom'
    )

# Show plot
plt.show()
```

### Task 4: Scatter Plot Comparisons Across Groups


#### 1. Scatter Plot for Compensation vs. Job Satisfaction by Employment Type


Visualize the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`), categorized by employment type (`Employment`). Use color coding or markers to differentiate between employment types.


```python
## Write your code here
# Query the data
query = """
SELECT ConvertedCompYearly, JobSatPoints_6, Employment
FROM main
WHERE ConvertedCompYearly IS NOT NULL AND JobSatPoints_6 IS NOT NULL AND Employment IS NOT NULL
"""
df = pd.read_sql_query(query, conn)

# Split 'Employment' column into individual categories if it contains multiple types
df['Employment'] = df['Employment'].str.split(',')

# Explode the employment types into separate rows
df_exploded = df.explode('Employment')

# Remove leading/trailing spaces from employment types
df_exploded['Employment'] = df_exploded['Employment'].str.strip()

# Remove rows with NaN values
df_cleaned = df_exploded.dropna(subset=['ConvertedCompYearly', 'JobSatPoints_6', 'Employment'])

# Create a scatter plot
plt.figure(figsize=(14, 8))
sns.scatterplot(
    data=df_cleaned,
    x='ConvertedCompYearly',
    y='JobSatPoints_6',
    hue='Employment',  # Different colors for each employment type
    palette='Set2',
    alpha=0.8
)

# Customize plot
plt.title('Scatter Plot: Compensation vs. Job Satisfaction by Employment Type')
plt.xlabel('Yearly Compensation (ConvertedCompYearly)')
plt.ylabel('Job Satisfaction (JobSatPoints_6)')
plt.legend(title='Employment Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)

# Show plot
plt.tight_layout()
plt.show()
```

#### 2. Scatter Plot for Work Experience vs. Age Group by Country


Compare work experience (`YearsCodePro`) across different age groups (`Age`) and countries (`Country`). Use colors to represent different countries and markers for age groups.


```python
## Write your code here

# Query the data
query = """
SELECT YearsCodePro, Age, Country
FROM main
WHERE YearsCodePro IS NOT NULL AND Age IS NOT NULL AND Country IS NOT NULL
"""
df = pd.read_sql_query(query, conn)

# Convert YearsCodePro to numeric (handle "Less than 1 year" and "More than 50 years")
def convert_years_code_pro(value):
    if "Less than 1 year" in value:
        return 0.5
    elif "More than 50 years" in value:
        return 50
    else:
        return pd.to_numeric(value, errors='coerce')

df['YearsCodePro'] = df['YearsCodePro'].apply(convert_years_code_pro)

# Remove rows with invalid YearsCodePro or Age values
df_cleaned = df.dropna(subset=['YearsCodePro', 'Age', 'Country'])

# Select top 5 countries for clarity
top_countries = df_cleaned['Country'].value_counts().head(5).index
df_filtered = df_cleaned[df_cleaned['Country'].isin(top_countries)]

# Create scatter plot
plt.figure(figsize=(14, 8))
sns.scatterplot(
    data=df_filtered,
    x='Age',
    y='YearsCodePro',
    hue='Country',  # Color-coded by country
    style='Age',    # Different markers for age groups
    palette='Set1',
    alpha=0.8
)

# Customize plot
plt.title('Scatter Plot: Work Experience vs. Age Group by Country')
plt.xlabel('Age Group')
plt.ylabel('Years of Professional Coding Experience (YearsCodePro)')
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)

# Show plot
plt.tight_layout()
plt.show()
```
