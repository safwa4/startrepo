### Task 1: Trends in Compensation Over Age Groups


##### 1. Line Chart of Median `ConvertedCompYearly` by Age Group


- Track how the median yearly compensation (ConvertedCompYearly) changes across different age groups.

- Use a line chart to visualize these trends.


```python
## Write your code here
# Query the data
query = """
SELECT Age, ConvertedCompYearly
FROM main
WHERE Age IS NOT NULL AND ConvertedCompYearly IS NOT NULL
"""
df = pd.read_sql_query(query, conn)

# Group data by Age and calculate the median compensation
median_compensation = df.groupby('Age')['ConvertedCompYearly'].median().reset_index()

# Sort by Age for proper ordering
median_compensation = median_compensation.sort_values(by='Age')

# Create the line chart
plt.figure(figsize=(12, 6))
plt.plot(median_compensation['Age'], median_compensation['ConvertedCompYearly'], marker='o', color='b', label='Median Compensation')

# Customize the plot
plt.title('Trends in Median Compensation Across Age Groups', fontsize=14)
plt.xlabel('Age Group', fontsize=12)
plt.ylabel('Median Compensation (Yearly)', fontsize=12)
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for readability
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

##### 2. Line Chart of Median `ConvertedCompYearly` for Ages 25 to 45


For a closer look, plot a line chart focusing on the median compensation for respondents between ages 25 and 45.


```python
## Write your code here

# Query the data
query = """
SELECT Age, ConvertedCompYearly
FROM main
WHERE Age IS NOT NULL AND ConvertedCompYearly IS NOT NULL
"""
df = pd.read_sql_query(query, conn)

# Filter relevant age groups
df_filtered = df[df['Age'].isin(['25-34 years old', '35-44 years old'])]

# Group data by Age range and calculate the median compensation
median_compensation = df_filtered.groupby('Age')['ConvertedCompYearly'].median().reset_index()

# Sort the age groups for proper ordering
age_order = ['25-34 years old', '35-44 years old']
median_compensation['Age'] = pd.Categorical(median_compensation['Age'], categories=age_order, ordered=True)
median_compensation = median_compensation.sort_values('Age')

# Create the line chart
plt.figure(figsize=(10, 6))
plt.plot(median_compensation['Age'], median_compensation['ConvertedCompYearly'], marker='o', color='b', label='Median Compensation')

# Customize the plot
plt.title('Median Compensation for Age Groups 25-34 and 35-44', fontsize=14)
plt.xlabel('Age Groups', fontsize=12)
plt.ylabel('Median Compensation (Yearly)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()

```

### Task 2: Trends in Job Satisfaction by Experience Level



##### 1. Line Chart of Job Satisfaction (`JobSatPoints_6`) by Experience Level



- Use a column that approximates experience level to analyze how job satisfaction changes with experience.

- If needed, substitute an available experience-related column for `Experience`.


```python
## Write your code here
# Query the data
query = """
SELECT YearsCodePro, JobSatPoints_6
FROM main
WHERE YearsCodePro IS NOT NULL AND JobSatPoints_6 IS NOT NULL
"""
df = pd.read_sql_query(query, conn)

# Clean and convert the YearsCodePro column
# Replace text values with numerical approximations
df['YearsCodePro'] = df['YearsCodePro'].replace({'Less than 1 year': 0.5, 'More than 50 years': 51})
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')

# Group by experience level (YearsCodePro) and calculate mean job satisfaction
experience_job_sat = df.groupby('YearsCodePro')['JobSatPoints_6'].mean().reset_index()

# Create the line chart
plt.figure(figsize=(12, 6))
plt.plot(experience_job_sat['YearsCodePro'], experience_job_sat['JobSatPoints_6'], marker='o', color='g', label='Job Satisfaction')

# Customize the plot
plt.title('Job Satisfaction by Experience Level', fontsize=14)
plt.xlabel('Years of Professional Coding Experience', fontsize=12)
plt.ylabel('Average Job Satisfaction (JobSatPoints_6)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

### Task 3: Trends in Job Satisfaction and Compensation by Experience


##### 1.Line Chart of Median ConvertedCompYearly Over Experience Level

- This line chart will track how median compensation (`ConvertedCompYearly`) changes with increasing experience.

- Use a column such as `WorkExp` or another relevant experience-related column.


```python
## Write your code here

# Query the data
query = """
SELECT WorkExp, ConvertedCompYearly
FROM main
WHERE WorkExp IS NOT NULL AND ConvertedCompYearly IS NOT NULL
"""
df = pd.read_sql_query(query, conn)

# Clean and convert the WorkExp column
# Replace text values with numerical approximations if needed
df['WorkExp'] = df['WorkExp'].replace({'Less than 1 year': 0.5, 'More than 50 years': 51})
df['WorkExp'] = pd.to_numeric(df['WorkExp'], errors='coerce')

# Group by WorkExp and calculate the median compensation
workexp_comp = df.groupby('WorkExp')['ConvertedCompYearly'].median().reset_index()

# Create the line chart
plt.figure(figsize=(12, 6))
plt.plot(workexp_comp['WorkExp'], workexp_comp['ConvertedCompYearly'], marker='o', color='g', label='Median Compensation')

# Customize the plot
plt.title('Median Compensation by Work Experience', fontsize=14)
plt.xlabel('Years of Work Experience', fontsize=12)
plt.ylabel('Median Yearly Compensation (USD)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

##### 2.Line Chart of Job Satisfaction (`JobSatPoints_6`) Across Experience Levels

- Create a line chart to explore trends in job satisfaction (`JobSatPoints_6`) based on experience level.

- This chart will provide insight into how satisfaction correlates with experience over time


```python
## Write your code here
# Query the data
query = """
SELECT WorkExp, JobSatPoints_6
FROM main
WHERE WorkExp IS NOT NULL AND JobSatPoints_6 IS NOT NULL
"""
df = pd.read_sql_query(query, conn)

# Clean and convert the WorkExp column
df['WorkExp'] = df['WorkExp'].replace({'Less than 1 year': 0.5, 'More than 50 years': 51})
df['WorkExp'] = pd.to_numeric(df['WorkExp'], errors='coerce')

# Group by WorkExp and calculate the mean Job Satisfaction
workexp_jobsat = df.groupby('WorkExp')['JobSatPoints_6'].mean().reset_index()

# Create the line chart
plt.figure(figsize=(12, 6))
plt.plot(workexp_jobsat['WorkExp'], workexp_jobsat['JobSatPoints_6'], marker='o', color='b', label='Job Satisfaction')

# Customize the plot
plt.title('Job Satisfaction Across Experience Levels', fontsize=14)
plt.xlabel('Years of Work Experience', fontsize=12)
plt.ylabel('Average Job Satisfaction (JobSatPoints_6)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
