### Task 1: Visualizing Data Composition with Pie Charts


##### 1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With


In the survey data, the `DatabaseWantToWorkWith` column lists the databases that respondents wish to work with. Let’s visualize the top 5 most-desired databases in a pie chart.



```python
##Write your code here

# Query the data
query = """
SELECT DatabaseWantToWorkWith
FROM main
WHERE DatabaseWantToWorkWith IS NOT NULL
"""
df = pd.read_sql_query(query, conn)

# Split the 'DatabaseWantToWorkWith' column into individual databases
df_expanded = df.assign(DatabaseWantToWorkWith=df['DatabaseWantToWorkWith'].str.split(';')).explode('DatabaseWantToWorkWith')

# Count the frequency of each database
database_counts = df_expanded['DatabaseWantToWorkWith'].value_counts()

# Select the top 5 databases
top_5_databases = database_counts.head(5)

# Add an "Other" category for remaining databases
other_count = database_counts.iloc[5:].sum()
top_5_databases['Other'] = other_count

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    top_5_databases,
    labels=top_5_databases.index,
    autopct='%1.1f%%',  # Show percentages
    startangle=140,  # Start angle for better aesthetics
    colors=plt.cm.Paired.colors,  # Use a colormap for distinct colors
)

# Add a title
plt.title('Top 5 Databases Respondents Want to Work With')

# Display the plot
plt.show()
```

The `DevType` column lists the developer types for respondents. We’ll examine the distribution by showing the top 5 developer roles in a pie chart.



```python
##Write your code here

# Query the data
query = """
SELECT DevType
FROM main
WHERE DevType IS NOT NULL
"""
df = pd.read_sql_query(query, conn)

# Split the 'DevType' column into individual developer roles
df_expanded = df.assign(DevType=df['DevType'].str.split(';')).explode('DevType')

# Count the frequency of each developer role
devtype_counts = df_expanded['DevType'].value_counts()

# Select the top 5 developer roles
top_5_devtypes = devtype_counts.head(5)

# Add an "Other" category for remaining developer roles
other_count = devtype_counts.iloc[5:].sum()
top_5_devtypes['Other'] = other_count

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    top_5_devtypes,
    labels=top_5_devtypes.index,
    autopct='%1.1f%%',  # Show percentages
    startangle=140,  # Start angle for better aesthetics
    colors=plt.cm.tab10.colors,  # Use a colormap for distinct colors
)

# Add a title
plt.title('Top 5 Developer Roles Among Respondents')

# Display the plot
plt.show()
```

##### 1.3 Create a pie chart for the operating systems used by respondents for professional use


The `OpSysProfessional` use column shows the operating systems developers use professionally. Let’s visualize the distribution of the top operating systems in a pie chart.



```python
##Write your code here

# Query the data with backticks around the column name containing spaces
query = """
SELECT `OpSysProfessional use`
FROM main
WHERE `OpSysProfessional use` IS NOT NULL
"""
df = pd.read_sql_query(query, conn)

# Rename the column 'OpSysProfessional use' to 'OpSysProfessional_use'
df = df.rename(columns={'OpSysProfessional use': 'OpSysProfessional_use'})

# Split 'OpSysProfessional_use' into individual operating systems if delimited (e.g., semicolons)
df_expanded = df.assign(OpSysProfessional_use=df['OpSysProfessional_use'].str.split(';')).explode('OpSysProfessional_use')

# Count the frequency of each operating system
os_counts = df_expanded['OpSysProfessional_use'].value_counts()

# Get the top 5 operating systems
top_os = os_counts.head(5)

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_os, labels=top_os.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Top 5 Operating Systems Used Professionally')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Show the plot
plt.show()
```
